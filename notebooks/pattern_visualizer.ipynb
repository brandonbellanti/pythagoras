```python
pattern = "5.-1.-2.-2"
```

```python
def plot_pattern(string):

    str_intervals = string.split('.') # split the string at the dots 

    int_intervals = list(map(int, str_intervals)) # convert str to int

    steps = []
    step = 0 # initial step
    steps.append(step) # add initial step to list
    
    for interval in int_intervals:
        step += interval 
        steps.append(step)
    
    print(steps)
    notes = [note_dict[step] for step in steps]
    print(notes)
    plt.plot(steps, 's', markersize=15)

    for count, step in enumerate(steps):
        plt.text(count-.2 ,step-.2, notes[count], color='w', weight='bold')
        
plot_pattern(pattern)
```

```python
notes = [note_dict[step] for step in steps]
notes
```

```python
plot_pattern(pattern)   
```

```python
for pattern in patterns:
    plot_pattern(pattern)
    plt.show()
```
